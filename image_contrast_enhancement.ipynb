##  Processing images

````python
img = cv.imread("me-at-smk.jpeg")
result = cv.convertScaleAbs(img, alpha=1.8, beta=30)
````

**cv2.convertScaleAbs( ) function** <br>
functionality : Scales, calculates absolute values, and converts the result to 8-bit. The parameters for the `cv2.convertScaleAbs( )` function are as follows: 

```cv.calcHist(image_in, alpha_in, beta_in)```

*   image_in – Input image.
*   alpha_in – Optional scale factor.
*   beta_in – Optional delta added to the scaled values.



## Showing image

````python
final_frame = cv.hconcat((img, result))
cv2_imshow(final_frame)
````
